```python
from __future__ import division

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 19.0.1 Data ImportÂ¶
We will now import the dataset.

```python
# load the master_dataset.xls
df = pd.read_excel('master_dataset.xlsx', sheetname='Sheet1')
# Print the `digits` data 
print(df.head())
```

```python
# Get the keys of the `digits` data
print(df.keys())
```

```python
target=df['Type'].values
```

```python
# Drop "Markdown1"
df.drop(['Type'],  axis = 1, inplace=True)
```

```python
# Drop Non-numeric cells
df.dropna()
```

```python
# Drop "Markdown1"
df.drop(['Date'],  axis = 1, inplace=True)
```

```python
data = df.values
```

```python
# Inspect the shape
print(data.shape)
```

```python
# Inspect the shape
print(target.shape)
```

```python
# Print the number of unique labels
number_digits = len(np.unique(target))
```

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Figure size (width, height) in inches
fig = plt.figure(figsize=(6, 6))

# Adjust the subplots 
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# For each of the 93 features
for i in range(93):
    # Initialize the subplots: add a subplot in the grid of 8 by 8, at the i+1-th position
    ax = fig.add_subplot(10, 10, i + 1, xticks=[], yticks=[])
    # Display an image at the i-th position
    ax.plot(data[i])#, interpolation='nearest')
    # label the image with the target value
    ax.text(0, 7, str(target[i]))

# Show the plot
plt.show()
```

```python
from sklearn.decomposition import RandomizedPCA
# Create a Randomized PCA model that takes two components
randomized_pca = RandomizedPCA(n_components=2)
```

```python
# Fit and transform the data to the model
reduced_data_rpca = randomized_pca.fit_transform(data)
```

```python
reduced_data_rpca
```

```python
from sklearn.decomposition import PCA
# Create a regular PCA model 
pca = PCA(n_components=2)
```

```python
# Fit and transform the data to the model
reduced_data_pca = pca.fit_transform(data)
```

```python
# Inspect the shape
reduced_data_pca.shape
```

```python
# Print out the data
print(reduced_data_rpca)
print(reduced_data_pca)
```

```python
colors = ['black', 'blue', 'purple', 'yellow', 'white', 'red', 'lime', 'cyan', 'orange', 'gray']
for i in range(len(colors)):
    x = reduced_data_rpca[:, 0][target == i]
    y = reduced_data_rpca[:, 1][target == i]
    plt.scatter(x, y, c=colors[i])
#plt.legend(target), bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title("PCA Scatter Plot")
plt.show()
```
