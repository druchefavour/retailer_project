# Predicting the impact of Markdowns on Weekly Sales in a Worldclass Retailer

## Introduction

In this project, we will be using data sets that reports the weekly sales and other vital information for a retail giant in the United States. The dataset include data on holidays and select major events that come up once a year for each department in the reatil store. 

It is the objective of the project to anaylise the data set so as to enable store executives make strategic decisions which would ultimately affect the bottom line. It is often suggested that markdowns do affect sales, so this project also sets out to test it as a hypohtesis and subsequently predict which departments will be affected and to what extent.
