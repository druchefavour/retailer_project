#### 12.0.3 Fitting a Classifier 
We are goint to fit Linear SVC classifier to the data. 

```python
# import the required libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.svm import SVC
```

```python
from sklearn import svm
# Set classifier =svm.LinearSVC and assign it lin_clf
lin_clf = svm.LinearSVC()
```

```python
# fit the classifier to the entire data set
lin_clf.fit(df_sales.values, df_targetIsHol.values)
```

#### 12.0.4 Measuring the Model's Performance
In this section we will compute the accuracy of the model. Let us split the data into two: 

* a training set
* a test set

We will fit/train the classifier on the training set, then we will make predictions on the test set. And then we will compare the prediction with the known labels by computing the accuracy of the model.


```python
# randomly split the data so that the test_size will be 1/3 of the initial sample size
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
                                                    random_state=21, stratify=y)
```

```python
#fit the classifier to the X_train and y_train
lin_clf.fit(X_train, y_train)
```

### 12.0.4 Prediction 
We will run prediction on the X_test and store the result as y_pred

```python
y_pred = lin_clf.predict(X_test)
```

```python
# print the test set
print("Test set predictions: \n {}".format(y_pred))
```

### 12.0.5 Model Accuracy
To compute the accuracy of the model we use score:

```python
lin_clf.score(X_test, y_test)
```
