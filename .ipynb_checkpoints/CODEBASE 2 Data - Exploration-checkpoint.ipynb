### 2.4.3 Relationship between Environmental Factors and Key Departmental Sales

In this section, we will explore the relationship between the selected 5 gift items of section 2.4.2 and the following environmental variables:

1. Temperature
2. Consumer Price Index (CPI) and 
3. Unemployment.

The selected items are as follows:

Jewelry
Cell_Phones
Floral
Clearance_Clothings
Video_Games

```python
# Create dataframe that contains the 8 features above
df_env_dep_sales = df_master[['Temperature','CPI', 'Unemployment', 'Jewelry', 'Cell_Phones', 'Floral', 
                            'Clearance_Clothings', 'Video_Games', 'IsHoliday']]
```

```python
df_env_dep_sales_corr_by_type = df_env_dep_sales.corr()
print(df_env_dep_sales_corr_by_type)
```

```python
fig, ax = plt.subplots(figsize=(10,8)) 
sns.heatmap(df_env_dep_sales_corr_by_type, 
            xticklabels=df_env_dep_sales_corr_by_type.columns.values,
            yticklabels=df_env_dep_sales_corr_by_type.columns.values)
```

As consumer price index (CPI) is a measure of changes in the price level of market basket of consumer goods and services purchased by households, it suggests that as CPI increases, households purchasing power increases. This is consistent with our result that shows that CPI is positively correlated to sales of gift items and especially cell-phone. As the spending power of households increases, so does the increase in sales of consumer items asuch as cell-phone.

As expected there is negative correlation between unemployment and CPI.

### 2.4.4 Relationship between Promo and Key Departmental Sales

Let us now investigate the relationship between the selected 5 gift items of section 2.4.2 and the promos: This will give us early indication on how the markdowns affect consumers behavior (in terms of purchase and sales)

```python
# Create dataframe that contains the 10 features above
df_promo_dep_sales = df_master[['MarkDown1','MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5', 'Jewelry', 'Cell_Phones', 'Floral', 
                            'Clearance_Clothings', 'Video_Games']]
```

```python
df_promo_dep_sales_corr_by_type = df_promo_dep_sales.corr()
print(df_promo_dep_sales_corr_by_type)
```

```python
fig, ax = plt.subplots(figsize=(10,8)) 
sns.heatmap(df_promo_dep_sales_corr_by_type, 
            xticklabels=df_promo_dep_sales_corr_by_type.columns.values,
            yticklabels=df_promo_dep_sales_corr_by_type.columns.values)
```

1. Strong positive correlation between MarkDown 1 and MarkDown 4 - Suggesting a spill over of customer behavior from Thanksgiving to superbowl.
2. Moderate positive correlation between MarkDown 1 and MarkDown 2 - This is expected as Superbowl usually takes place in February while MarkDown 3 (Easter) takes place between March and April. 
3. With respect to Departmental gift items, the relationship with the MarkDowns is insignificant except an observed weak negative correlation between MarkDown1, MarkDown 2 and MarkDown 3 and the purchase of clearance clothing. This suggests that customers would tend to purchase clearance clothing with or without the promos and infact sales of clearance clothing may be negatively impacted by the promos.
